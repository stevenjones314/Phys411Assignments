## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200000
x = np.random.normal(0,1,N)
x = x-np.mean(x)

a = 10.0;
n = np.logspace(0.01,100,40)
r = np.zeros(40)
noise_inv = np.zeros(40)

for i in range(40):
    noise = np.random.normal(0,n[i],N)
    y = a*x + noise
    y = y-np.mean(y)
    r[i] = np.sum(x*y)/np.std(x,ddof=1)/np.std(y,ddof=1)/(N-1)
    noise_inv[i] = np.std(a*x/noise)
    
  
fig=figure(figsize=(8, 6));ax=fig.add_subplot(2,1,1);
ax.plot(noise_inv,r,label='Linear Plot')
ax.set_xlabel(''),ax.set_ylabel('Correlation Coefficients')
ax.set_title('Relationship Between Correlation Coefficeint and Noise')
hl=ax.legend(loc=4,fontsize='small')

ax=fig.add_subplot(2,1,2)
ax.loglog(noise_inv,r,label='Log Plot')
ax.set_xlabel('Normalized Inverse of Noise'),ax.set_ylabel('Correlation Coefficients')
hl=ax.legend(loc=4,fontsize='small')


# This comparison shows that when the distribution of noise of the linearly dependent variable is small compared
# compared to the amplitude of the independent variable the correlation coefficient will be larger
# i.e. when the standard deviation of the inverse of the noise normalized by the signal is large (small relative noise)
# the correlation coefficient is large(er)
# when the distribution of the noise is large compared to the independent variable, then the correlation coeficient will
# vary dependent on the actual values of the introduced noise
```

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
a = 1.0
n = 1.0
b = np.zeros(10000)
s_xy = np.zeros(10000)

for i in range(10000):
    x = np.random.normal(0,1,N); # x = x-np.mean(x)
    y = a*x + np.random.normal(0,n,N); # y=y-np.mean(y)
    p = np.polyfit(x,y,1)
    b[i] = p[0]
    s_xy[i] = np.sqrt(np.sum((y-np.mean(y))*(y-np.mean(y)))/(N-2))


fig=figure(figsize=(12, 6));ax=fig.add_subplot(1,2,1); 
ax.hist(b,30)
ax.set_xlabel('Slope (b)');ax.set_ylabel('Occurences');
ax.set_title('Slope of Linear Fit')

ax=fig.add_subplot(1,2,2)
ax.hist(s_xy,30)
ax.set_xlabel('s_xy');ax.set_ylabel('Occurences');
ax.set_title('Standard Deviation of xy')


# The first histogram 'Slope of Linear Fit' has a similar shape to a normal distribution which confirms the hypothesis that
# the slope will have a student-t distribution which is essentially a normal distribution for a small sample size
# The second histogram 'Standard deviation of xy' also shows a 'normal' distribution shape.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

N = 10
a = 1.0
n = 1.0
b = np.zeros(10000)
s_xy = np.zeros(10000)

for i in range(10000):
    x = np.random.normal(0,1,N); # x = x-np.mean(x)
    y = a*x + np.random.normal(0,n,N); # y=y-np.mean(y)
    p = np.polyfit(x,y,1)
    b[i] = p[0]
    s_xy[i] = np.sqrt(np.sum((y-np.mean(y))*(y-np.mean(y)))/(N-2))


fig=figure(figsize=(12, 6));ax=fig.add_subplot(1,1,1); 
ax.hist(b,30,label='test data')
ax.set_xlabel('Slope (b)');ax.set_ylabel('Occurences')
ax.set_title('Slope of Linear Fit')

x_studentt = np.linspace(-2,3,50)
y_studentt = 2500*mlab.normpdf(x_studentt, 1., 1.)
ax.plot(x_studentt,y_studentt,'k--',label = 'Student t Dist.')
hl=ax.legend(loc=1)
```

**NOTE** There is a way to treat both axes independently called "Neutral Regression".  It is relatively simple, and involves normalizing the total distance of each point from the trend line. <a href="http://journals.ametsoc.org/doi/pdf/10.1175/1520-0426(1999)016%3C0876%3AAPFANR%3E2.0.CO%3B2">(Marsden 1999)</a>. This technique is strongly preferred if there is no reason to think that one data set is dependent on the other.  A caveat is that the data sets shoudl be normalized before fitting so that the "distance" has meaning.
