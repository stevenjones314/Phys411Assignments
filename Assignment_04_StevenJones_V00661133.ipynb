1 Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization. The number of data points N can be relatively small (say 1000).

```python
import numpy as np

def cum_sum(N=1000,sigma=1,mu=0,plt=0):
    #N = 1000
    x = sigma*np.random.randn(N)+mu
    x_sum = np.zeros(N)

    for i in range(N):
        x_sum[i] = np.sum(x[0:i])
    if plt == 1:
        fig, ax=subplots(1,1)
        ax.plot(x, label='Normally Distributed Deta')
        ax.plot(x_sum, label = 'Cumulative Sum')
        ax.legend(loc=0)
        ax.set_xlabel('Data Point')
        ax.set_ylabel('Value')
        ax.set_title('"Red" Time Series')
    
    return(x_sum)
    
cum_sum(1000, 100, -5,plt = 1);
```

2 Write a second function to return the estimate of the power spectrum of x. For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate. Plot the resulting spectrum.

```python
import scipy.stats as stats

def est_ps(x,dt=1.,plt=0):
    N = shape(x)
    N = N[0]
    T = dt*N
    f = np.arange(N/2)/T
    X = np.fft.fft(x)
    Gxx = np.real(0.5*T*np.conj(X[:N/2])*X[:N/2])
    if plt == 1:
        fig,ax=subplots(1,1)
        ax.loglog(f[1:],Gxx[1:])
        ax.set_xlabel('f [Hz]')
        ax.set_ylabel('$G_{xx}(f_l)$')
        ax.set_title('Power Spectrum')
    return Gxx, f
x = cum_sum(1000, 100,-5,plt = 1)
[Gxx,f] = est_ps(x,plt = 1);
```

4 Now, on the same plot, add the mean of all your Gxx(f) estimates (as a function of f), and the upper and lower bounds encompasing 95% of the data. (Hint, sort each column of your matrix - np.sort - and plot the value at 0.025N and 0.975N).

```python
import matplotlib.pyplot as plt
N = 100
Gxs = np.zeros((N,500))
fig, ax=subplots(1,1)



for i in range(N):
    x = cum_sum(1000)
    [Gxs[i,:],f] = est_ps(x)
    ax.loglog(f,Gxs[i,:],'+')

Gxs_mean = np.mean(Gxs,0)
Gxs_sort = np.sort(Gxs,0)

ax.loglog(f,Gxs_mean,label='Mean')
ax.loglog(f,Gxs_sort[N*0.025,:],label='Lower Bound')
ax.loglog(f,Gxs_sort[N*0.975,:],label='Upper Bound')
ax.legend()
ax.set_xlabel('f [Hz]')
ax.set_ylabel('$G_{xx}(f)$')
ax.set_title('Power Spectra for N = %d data sets' %N )
```
