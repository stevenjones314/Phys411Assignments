## Error bounds on coherence estimates

The 95% confidence intervals for the coherence squared $γ^2_{xy}(f)$ between two variables x(t) and y(t) are approximately:

$
\gamma_{xy}^2 \pm \frac{\sqrt{2}\left(1+\gamma_{xy}^2\right)}{|\gamma_{xy}^2|\sqrt{n_d}}
$

where $n_d$ is the number of blocks used to calculate the coherence.
You may use matplotlib.mlab.csd for the following question, but you had best use it correctly!

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats

Num = 10000
t = arange(Num)
x = 1.*np.random.randn(Num)
n = 0.1*np.random.randn(Num)
y = np.zeros(Num)
b = 10

for ind in range(Num):
    if(ind>b):
        a = b*1.
        y[ind] = n[ind] + np.sum(x[ind-b:ind])/a
    else:
        a = ind*1.
        if(ind==0):
            a = 1
        y[ind] = n[ind] + np.sum(x[:ind])/a
    
fig,ax=plt.subplots(1,1,figsize=(8,6)) 
ax.plot(t,x,label='x')
ax.plot(t,y,label='y')
ax.legend(loc='best')
ax.set_xlabel('Time')
ax.set_ylabel('Val')

fig,ax=plt.subplots(1,1,figsize=(8,6)) 
ax.plot(t,x,label='x')
ax.plot(t,y,label='y')
ax.legend(loc='best')
ax.set_xlabel('Time')
ax.set_ylabel('Val')
ax.set_title('Zoomed')
xlim((0,b*30))
```
