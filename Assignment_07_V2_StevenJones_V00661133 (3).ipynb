1 Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of f c =0.06 Hz .

1.plot the signal and the smoothed signal (suggest zooming in on the step).

2.plot the spectral response |H(f)| 

3.plot the phase response angle(H(f)) .

Be sure to comment on how good your filter is. 

```python
fc = 0.06
fs = 1./dt
df = fs/N
ff = np.linspace(0.,fs,N)

X = np.fft.fft(xn)
#to filter, we zero the data points that correpond to frequencies above the cut off frequency
#note, because the fourier transform is symmetric and the last half of the data is a copy of the first half
#we must symmetrically remove these point about the middle point
X[(ff>fc) & (ff<(fs-fc))] = 1.e-21
y = real(np.fft.ifft(X))+np.mean(xn)
py,fy = mlab.psd(y,**args)


fig,ax=plt.subplots(4,1,figsize=(8,14)) 
ax[0].plot(t,x,label='Sine wave')
#ax[0].plot(t,xj,label='Signal - sine wave with jump')
ax[0].plot(t,xn,label='Signal w/ noise')
ax[0].plot(t,y,label='Filtered Signal')
ax[0].legend(loc='best')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Val')


ax[1].loglog(f,pn,'--',label="Noise");
ax[1].loglog(f,px,linewidth=1.2,label="Sine")
ax[1].loglog(f,pxn,label='Signal + Noise')
ax[1].loglog(fy,py,label='Filtered Signal')
ax[1].set_xlabel('f [Hz]')
ax[1].set_ylabel(r'$G_{xx} \mathrm{[V^2 Hz^{-1}]}$')
ax[1].legend(fontsize='small',loc='best')


ax[2].loglog(fy,sqrt(py/(pxn)),'k',label = 'Filter Amplitude Response')
#ax[2].set_ylim((1.e-2,4))
ax[2].set_ylabel(r'$|H(f)|$')
ax[2].legend(loc='best')

Gxy,f_csd  = mlab.csd(xn,y,Fs=1./dt,NFFT=nfft)

ax[3].semilogx(f_csd,angle(Gxy),'k',label = 'Filter Phase Response')
ax[3].set_ylabel('phase')
ax[3].set_xlabel('f [Hz]')
ax[3].legend(loc='best')
```

2 Now repeat the above but apply a Hanning window to the frequency window before you apply it to the FFT of x n  . Comment on how well the filter is doing when you do this. (Its helpful to keep the previous result up) 

```python
fc = 0.06
fs = 1./dt
df = fs/N
ff = np.linspace(0.,fs,N)

NN = 2*fc*N/fs
hanning = np.hanning(NN)
hanning_1 = hanning[:NN/2+1]
hanning_2 = hanning[NN/2+1:]



X_hann = np.fft.fft(xn)
X_hann[(ff<fc)] = X_hann[(ff<fc)]*hanning_2
X_hann[(ff>(fs-fc))] = X_hann[(ff>(fs-fc))]*hanning_1

X_hann[(ff>fc) & (ff<(fs-fc))] = 1.e-21


y_hann = real(np.fft.ifft(X_hann))+np.mean(xn)
py_hann,fy = mlab.psd(y_hann,**args)


fig,ax=plt.subplots(4,1,figsize=(8,14)) 
ax[0].plot(t,x,label='Sine wave')
ax[0].plot(t,xn,label='Signal w/ noise')
#ax[0].plot(t,xn_hann,label='Signal w/ hann window')
ax[0].plot(t,y_hann,label='Filtered Signal')
ax[0].legend(loc='best')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Val')


ax[1].loglog(f,pn,'--',label="Noise");
ax[1].loglog(f,px,linewidth=1.2,label="Sine")
ax[1].loglog(f,pxn,label='Signal + Noise')
ax[1].loglog(fy,py_hann,label='Filtered Signal')
ax[1].set_xlabel('f [Hz]')
ax[1].set_ylabel(r'$G_{xx} \mathrm{[V^2 Hz^{-1}]}$')
ax[1].legend(fontsize='small',loc='best')


ax[2].loglog(fy,sqrt(py_hann/(pxn)),'k',label = 'Filter Amplitude Response')
#ax[2].set_ylim((1.e-2,4))
ax[2].set_ylabel(r'$|H(f)|$')
ax[2].legend(loc='best')

Gxy_hann,f_csd  = mlab.csd(xn,y_hann,Fs=1./dt,NFFT=nfft)

ax[3].semilogx(f_csd,angle(Gxy_hann),'k',label = 'Filter Phase Response')
ax[3].set_ylabel('phase')
ax[3].set_xlabel('f [Hz]')
ax[3].legend(loc='best')
```

Here we can see that the hanning window causes the frequency to attenuate much more slowly than before, which causes attentuation of the desired signal peak.

```python
fc = 0.09
fs = 1./dt
df = fs/N
ff = np.linspace(0.,fs,N)

NN = 2*fc*N/fs
hanning = np.hanning(NN)
hanning_1 = hanning[:NN/2]
hanning_2 = hanning[NN/2+1:]



X_hann = np.fft.fft(xn)
X_hann[(ff<fc)] = X_hann[(ff<fc)]*hanning_2
X_hann[(ff>(fs-fc))] = X_hann[(ff>(fs-fc))]*hanning_1

X_hann[(ff>fc) & (ff<(fs-fc))] = 1.e-21


y_hann = real(np.fft.ifft(X_hann))+np.mean(xn)
py_hann,fy = mlab.psd(y_hann,**args)


fig,ax=plt.subplots(4,1,figsize=(8,14)) 
ax[0].plot(t,x,label='Sine wave')
ax[0].plot(t,xn,label='Signal w/ noise')
#ax[0].plot(t,xn_hann,label='Signal w/ hann window')
ax[0].plot(t,y_hann,label='Filtered Signal')
ax[0].legend(loc='best')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Val')


ax[1].loglog(f,pn,'--',label="Noise");
ax[1].loglog(f,px,linewidth=1.2,label="Sine")
ax[1].loglog(f,pxn,label='Signal + Noise')
ax[1].loglog(fy,py_hann,label='Filtered Signal')
ax[1].set_xlabel('f [Hz]')
ax[1].set_ylabel(r'$G_{xx} \mathrm{[V^2 Hz^{-1}]}$')
ax[1].legend(fontsize='small',loc='best')


ax[2].loglog(fy,sqrt(py_hann/(pxn)),'k',label = 'Filter Amplitude Response')
#ax[2].set_ylim((1.e-2,4))
ax[2].set_ylabel(r'$|H(f)|$')
ax[2].legend(loc='best')

Gxy_hann,f_csd  = mlab.csd(xn,y_hann,Fs=1./dt,NFFT=nfft)

ax[3].semilogx(f_csd,angle(Gxy_hann),'k',label = 'Filter Phase Response')
ax[3].set_ylabel('phase')
ax[3].set_xlabel('f [Hz]')
ax[3].legend(loc='best')
```

## Q2: Implementing a bandpass filter

1 Create a red-noise process with 20000 data points, dt=1 s . Using a truncated Dirichlet Kernel, create a bandpass filter centered at f c =0.1 Hz  that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz. 

1.Plot the unshifted and shifted filters versus k , the time index.

HINT: when you make the Dirichlet Kernel filter make sure that the k=0  case is treated "correctly" because sin(0)=0  and you will get a discontinuity. Calculate what h 0   should be from L'Hôpital's_rule.


```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab


N=20000
fc = 0.05
dt = 1.
fs = 1./dt
t=np.arange(0,N*dt,dt)

x = np.cumsum(randn(N)) # red noise
f0=0.03 # Hz
x = x+ 20.*sin(f0*2*pi*t) # + sine wave
x=x-np.mean(x)
noise = 20.*randn(N)
xn = x+noise
# remove the mean!!
xn = xn-np.mean(xn)
#
```

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab

# Calculating the truncated Dirichlet Kernel to filter at 0.05 Hz lowpass:
k = np.array(range(-N/2+1,N/2))
fc = 0.05
M = fc*2*N+1
h2 = (1./N)*np.sin(np.pi*(k+0.000001)*M/N)/np.sin(np.pi*(k+0.000001)/N)


K = 1001
h = 1.0*h2
h[abs(k)>(K-1)/2]=0
w = 0.*h
w[abs(k)<=(K-1)/2]=np.hanning(K)
H2 = N*fftshift(ifft(h*w))

#h2[abs(k)>(K-1)/2]=0
#H2 = N*fftshift(ifft(h2))

H2_shift1 = 1.e-15*H2
H2_shift1[N/2+1100-200:N/2+3*1100-200] = H2[N/2-1100:N/2+1100]
H2_shift2 = 1.e-15*H2
H2_shift2[N/2-(3*1100-200):N/2-(1100-200)] = H2[N/2-1100:N/2+1100]
H2_shift = H2_shift1+H2_shift2

#h2_shift = (np.fft.fftshift(np.fft.ifft(np.fft.fftshift(H2_shift))))
h2_shift = -1.*np.fft.ifft(np.fft.fftshift(H2_shift))

fig,ax=plt.subplots(1,1)
ax.plot(k,h2,label='unshifted')
ax.plot(k,h2_shift,label='shifted')
ax.set_xlabel('k')
ax.set_ylabel('$h_k$')
ax.set_xlim([-100,300])
ax.legend(loc='best')

fig,ax=plt.subplots(1,1)
ax.plot(k*1./N,abs(H2),label='unshifted')
ax.plot(k*1./N,abs(H2_shift),label='shifted')
ax.set_xlabel('m')
ax.set_ylabel('$h_k$')
ax.legend(loc='best')
#ax.set_xlim([-2000,2000])


filtlen = 200
h = h2_shift*1.
# how long is the filter:
h = h[abs(k)<filtlen+1]

y = np.convolve(xn,h,mode='same')
fig,ax=plt.subplots(4,1,figsize=(5,7))
ax[0].plot(xn-np.mean(xn),label='Raw')
ax[0].plot(y,label='Filtered',color='r')
ax[0].set_xlim([500,1000])
ax[0].set_xlabel('t [s]');ax[0].set_ylabel('x [V]');ax[0].legend()
ax[0].legend(loc='best')

nfft=256
pxx,f=mlab.psd(xn,Fs=1.,NFFT=nfft,noverlap=nfft/2)
fx = f*1.
ax[1].loglog(f,pxx,label='raw')
pyy,f=mlab.psd(y,Fs=1.,NFFT=nfft,noverlap=nfft/2)


ax[1].loglog(f,pyy,color='r',label='filtered')
ax[1].set_ylabel('$G_{xx}\ \ [V^2 Hz^{-1}]$')
ax[1].set_xlabel('$f\ [Hz]$')
ax[1].legend(loc='best')
ax[2].loglog(fx,pyy[127:]/pxx)
ax[2].set_ylabel('$H^2 = G_{yy}/G_{xx}$')
ax[2].set_xlabel('f [Hz]')
ax[2].legend(loc='best')
ax[0].set_title('Filter length: %d'%(2*filtlen+1))
plt.tight_layout()

Gxy,f_csd  = mlab.csd(xn,y,Fs=1./dt,NFFT=nfft)

ax[3].semilogx(f_csd,angle(Gxy),'k',label = 'Filter Phase Response')
ax[3].set_ylabel('phase')
ax[3].set_xlabel('f [Hz]')
```

We can see that the correct frequencies are blocked and passed
